```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
CRCT_Grade3_2013_url = 'https://raw.githubusercontent.com/tracyhua2/SYS3034-APSCase/refs/heads/main/Data/CRCT_Grade3_2013.csv'
Grade3_2013_df = pd.read_csv(CRCT_Grade3_2013_url)
Grade3_2013_df['Year'] = 2013
Grade3_2013_df.head()
```

```python
CRCT_Grade3_2014_url = 'https://raw.githubusercontent.com/tracyhua2/SYS3034-APSCase/refs/heads/main/Data/CRCT_Grade3_2014.csv'
Grade3_2014_df = pd.read_csv(CRCT_Grade3_2014_url)
Grade3_2014_df['Year'] = 2014
Grade3_2014_df.head()
```

```python
Grade3_2013_df.columns
```

```python
Grade3_2013_df.dtypes
```

```python
Grade3_2013_df['System name'] = Grade3_2013_df['System name'].astype(str)
Grade3_2013_df['School name'] = Grade3_2013_df['System name'].astype(str)
```

```python
# make everything float then fill NAs
exclude_cols = ['School name', 'System name']
cols_to_convert = [col for col in Grade3_2013_df.columns if col not in exclude_cols]
Grade3_2013_df[cols_to_convert] = Grade3_2013_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
# overall statistics
def statistics(df, subject):
  mean = df[subject].mean()
  max = df[subject].max()
  min = df[subject].min()
  std = df[subject].std()
  return mean, max, min, std
```

```python
# overall reading statistics
overall_reading = statistics(Grade3_2013_df,'Reading: Mean Scale Score')
```

```python
# overall ELA statistics
overall_ELA = statistics(Grade3_2013_df,'ELA: Mean Scale Score')
```

```python
# overall math statistics
overall_math = statistics(Grade3_2013_df,'Math: Mean Scale Score')
```

```python
# overall science statistics
overall_science = statistics(Grade3_2013_df,'Science: Mean Scale Score')
```

```python
# overall social studies statistics
overall_social = statistics(Grade3_2013_df,'Social Studies: Mean Scale Score')
```

```python
APS_Grade3_2013 = Grade3_2013_df[Grade3_2013_df['System name'].isin(['ATLANTA PUBLIC SCHOO'])]
APS_Grade3_2013.tail()
```

```python
# make everything float then fill NAs
exclude_cols = ['School name', 'System name']
cols_to_convert = [col for col in APS_Grade3_2013.columns if col not in exclude_cols]
APS_Grade3_2013[cols_to_convert] = APS_Grade3_2013[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
# reading score statistics
reading_mean = APS_Grade3_2013['Reading: Mean Scale Score'].mean()
reading_max = APS_Grade3_2013['Reading: Mean Scale Score'].max()
reading_min = APS_Grade3_2013['Reading: Mean Scale Score'].min()
reading_std = APS_Grade3_2013['Reading: Mean Scale Score'].std()

print(f'Mean: {reading_mean}, Range: {reading_min, reading_max}, STD: {reading_std}')
```

```python
# ELA score statistics
ELA_mean = APS_Grade3_2013['ELA: Mean Scale Score'].mean()
ELA_max = APS_Grade3_2013['ELA: Mean Scale Score'].max()
ELA_min = APS_Grade3_2013['ELA: Mean Scale Score'].min()
ELA_std = APS_Grade3_2013['ELA: Mean Scale Score'].std()

print(f'Mean: {ELA_mean}, Range: {ELA_min, ELA_max}, STD: {ELA_std}')
```

```python
# math score statistics
math_mean = APS_Grade3_2013['Math: Mean Scale Score'].mean()
math_max = APS_Grade3_2013['Math: Mean Scale Score'].max()
math_min = APS_Grade3_2013['Math: Mean Scale Score'].min()
math_std = APS_Grade3_2013['Math: Mean Scale Score'].std()

print(f'Mean: {math_mean}, Range: {math_min, math_max}, STD: {math_std}')
```

```python
# science score statistics
science_mean = APS_Grade3_2013['Science: Mean Scale Score'].mean()
science_max = APS_Grade3_2013['Science: Mean Scale Score'].max()
science_min = APS_Grade3_2013['Science: Mean Scale Score'].min()
science_std = APS_Grade3_2013['Science: Mean Scale Score'].std()

print(f'Mean: {science_mean}, Range: {science_min, science_max}, STD: {science_std}')
```

```python
# social science score statistics
social_mean = APS_Grade3_2013['Social Studies: Mean Scale Score'].mean()
social_max = APS_Grade3_2013['Social Studies: Mean Scale Score'].max()
social_min = APS_Grade3_2013['Social Studies: Mean Scale Score'].min()
social_std = APS_Grade3_2013['Social Studies: Mean Scale Score'].std()

print(f'Mean: {social_mean}, Range: {social_min, social_max}, STD: {social_std}')
```

```python
subjects = ['Reading','ELA','Math','Science','Social Studies']
means = [reading_mean, ELA_mean, math_mean, science_mean, social_mean]
overall_mean = [overall_reading[0], overall_ELA[0], overall_math[0], overall_science[0],
               overall_social[0]]
avg_overall_mean = np.mean(overall_mean)

plt.figure(figsize=(8,5))
plt.bar(subjects, means)
plt.axhline(avg_overall_mean, linestyle='dotted',color='red', label='Average of Overall Mean in Atlanta')
plt.title('Mean Scores by Subject in Atlanta Public Schools')
plt.xlabel('Subjects')
plt.ylabel('Mean Score')
plt.ylim(800,840)
plt.legend()
plt.show()
```

```python
std = [reading_std, ELA_std, math_std, science_std, social_std]
overall_std = [overall_reading[3], overall_ELA[3], overall_math[3], overall_science[3],
               overall_social[3]]
avg_overall_std = np.mean(overall_std)

plt.figure(figsize=(8,5))
plt.bar(subjects, std)
plt.axhline(avg_overall_std, linestyle='dotted',color='red', label='Average of Overall Standard Deviation in Atlanta')
plt.title('Standard Devation of Scores by Subject in Atlanta Public Schools')
plt.xlabel('Subjects')
plt.ylabel('Standard Devation of Score')
plt.ylim(0,30)
plt.legend()
plt.show()
```
